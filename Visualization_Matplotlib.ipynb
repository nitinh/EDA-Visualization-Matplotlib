```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
x=[1,2,3]
y=[2,4,6]

#Keyword Argument Notation
#plt.plot(x,y,label='2x',color='magenta',linewidth=3,linestyle='--',markersize=10)
#short hand notation
plt.figure(figsize=(5,3),dpi=300) # 5 inches by 3 inches image size
#Create line1
plt.plot(x,y,'mo--',label='2x') # Many other shorthand notations

# Line 2 
# Create data for lIne2
x2=np.arange(0,4,0.4)
print(x2)
#Create Line2
plt.plot(x2,x2**2,'r+-',label='x^2')


#Plot Title
plt.title('Test Line Graph',fontdict={'fontname' : 'Comic Sans MS', 'fontsize':20})

#X and Y Lables
plt.xlabel("Month")
plt.ylabel("Temperature")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()

#Save figure (dpi = 300 is good resolution when saving so graph has high resolution)
plt.savefig('testgraph.png')
```

```
#Bar Chart

labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values)
patterns=['/','o','*']
for bar in bars:
    #bar.set_color('red')
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('/')
#bars[1].set_hatch('o')
#bars[2].set_hatch('+#')
plt.figure(figsize=(6,4))
plt.show()
```
